```python
import pandas as pd
```

```python
data = pd.read_excel('survey_results.xlsx',sheet_name='All Data')
data.info()
```

```python
# replace Gender column with 1 and 0 
def replaceGender(value):
    if value == 'Male':
        return 1
    else:
        return 0

data['Gender'] = data['Gender'].map(replaceGender)
data
```

```python
relevant_data = data.select_dtypes(include=['float64'])
relevant_data
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
corr = relevant_data.corr()
corr.style.background_gradient(cmap='coolwarm')
#sns.heatmap(data.corr(),mask=np.zeros_like(data.corr(), dtype=np.bool), cmap=sns.diverging_palette(220, 10, as_cmap=True),square=True, ax=ax)
```
