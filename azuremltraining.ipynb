```python
# azureml-core of version 1.0.72 or higher is required
# azureml-dataprep[pandas] of version 1.1.34 or higher is required
from azureml.core import Workspace, Dataset
import pandas as pd

subscription_id = '0c872f26-5716-46e2-9c44-a92f5399706a'
resource_group = 'sptrainingresourcegroup'
workspace_name = 'sptrainingmlws'

workspace = Workspace(subscription_id, resource_group, workspace_name)

dataset = Dataset.get_by_name(workspace, name='sptrainingdataset')
pddataset = dataset.to_pandas_dataframe()
#pddataset.corr()
```

```python
pddataset.info()
```

```python
newdataset = pddataset.iloc[:,:12]
newdataset.info()
```

```python
newdataset.set_index('id',inplace=True)
```

```python
zscoredataset = newdataset.iloc[:,1:]
```

```python
zscoredataset.head()
```

```python
stats = zscoredataset.describe().T
stats
```

```python
zscoreddata = (zscoredataset - stats['mean'])/stats['std']
```

```python
zscoreddata['diagnosis'] = newdataset['diagnosis']
```

```python
display(zscoreddata)
```

```python
from azureml.core import Workspace, Datastore, Dataset

pandas_df = zscoreddata

#datastore = Datastore.get_default(workspace)
datastore = Datastore(workspace)
dataset = Dataset.Tabular.register_pandas_dataframe(pandas_df, datastore, "dataset_from_pandas_df", show_progress=True)
```
